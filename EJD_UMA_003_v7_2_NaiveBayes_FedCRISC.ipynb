## Objetivo

El Prof. Lopez Rubio senalo dos limitaciones criticas en la version anterior: (1) el conjunto NSL-KDD KDDTrain+ presenta un techo de F1 cercano a 0.99 que impide demostrar superioridad estadistica entre modelos, y (2) los pesos de agregacion asignados manualmente no tienen justificacion empirica. Esta version responde a ambas observaciones de forma directa.

Para el techo de F1 se incorpora KDDTest+21, el subconjunto de prueba del NSL-KDD que contiene ataques no presentes en el entrenamiento. Al exponer el modelo a condiciones fuera de distribucion, la ventaja de la mezcla se vuelve observable y estadisticamente medible.

Para los pesos arbitrarios se introduce un procedimiento de aprendizaje: se reserva el 20% del entrenamiento como conjunto de validacion, y los pesos de cada nodo se optimizan mediante busqueda directa para maximizar el F1-macro del modelo federado sobre ese conjunto. Los pesos resultantes tienen respaldo empirico y son reproducibles.

Las variables contextuales CRISC (ICC, CMM, KCI, KRI, CVSS) se mantienen como capa de interpretacion institucional que conecta el ejercicio con la linea de investigacion sobre infraestructura critica.

```python
# ╔══════════════════════════════════════════════════════════════╗
# ║  SECCION 3 · Funciones de particion, agregacion y prediccion ║
# ╚══════════════════════════════════════════════════════════════╝

def particion_dirichlet(X, y, n_nodos, alpha, semilla):
    rng = np.random.default_rng(semilla)
    clases_unicas = np.unique(y)
    indices_nodo  = [[] for _ in range(n_nodos)]
    for clase in clases_unicas:
        idx   = np.where(y == clase)[0]
        rng.shuffle(idx)
        props  = rng.dirichlet([alpha] * n_nodos)
        cortes = (props * len(idx)).astype(int)
        cortes[-1] = len(idx) - cortes[:-1].sum()
        cortes = np.maximum(cortes, 1)
        inicio = 0
        for i, n in enumerate(cortes):
            indices_nodo[i].extend(idx[inicio:inicio+n].tolist())
            inicio += n
    return [(X[idx], y[idx]) for idx in indices_nodo]


def entropia_local(y_nodo, n_clases):
    conteo = np.bincount(y_nodo, minlength=n_clases).astype(float)
    dist   = conteo / conteo.sum()
    return entropy(dist + 1e-10)


def agregar_nb(modelos, pesos, n_clases, n_rasgos, log_prior, modo='mezcla'):
    pesos_norm = np.array(pesos) / np.sum(pesos)
    mu_fed     = np.zeros((n_clases, n_rasgos))
    var_fed    = np.zeros((n_clases, n_rasgos))
    for modelo, w in zip(modelos, pesos_norm):
        for j, cid in enumerate(modelo.classes_):
            mu_k  = modelo.theta_[j]
            var_k = modelo.var_[j]
            if modo == 'mezcla':
                mu_fed[cid]  += w * mu_k
                var_fed[cid] += w * (var_k + mu_k**2)
            else:
                mu_fed[cid]  += w * mu_k
                var_fed[cid] += w * var_k
    if modo == 'mezcla':
        var_fed = var_fed - mu_fed**2
    var_fed = np.clip(var_fed, 1e-9, None)
    return mu_fed, var_fed, log_prior


def predecir_nb(X, mu, var, log_prior):
    lp = np.zeros((X.shape[0], mu.shape[0]))
    for c in range(mu.shape[0]):
        lp[:, c] = log_prior[c] - 0.5 * np.sum(
            np.log(2 * np.pi * var[c]) + (X - mu[c])**2 / var[c], axis=1
        )
    return np.argmax(lp, axis=1)


def aprender_pesos(modelos_locales, X_val, y_val, n_clases, n_rasgos, log_prior, semilla=42):
    """Aprende los pesos de cada nodo minimizando 1-F1 en el conjunto de validacion.
    Se agrega regularizacion L2 sobre los pesos para evitar soluciones degeneradas
    donde un solo nodo recibe peso=1 y los demas peso=0.
    El parametro lambda_reg controla el balance entre F1 y distribucion uniforme.
    """
    n       = len(modelos_locales)
    LAMBDA  = 0.05   # Penalizacion L2: evita colapso a un solo nodo

    def objetivo(w_raw):
        w = np.exp(w_raw)
        w = w / w.sum()
        mu, var, lp = agregar_nb(modelos_locales, w, n_clases, n_rasgos, log_prior, modo='mezcla')
        y_pred   = predecir_nb(X_val, mu, var, lp)
        f1       = f1_score(y_val, y_pred, average='macro', zero_division=0)
        reg_l2   = LAMBDA * np.sum((w - 1.0/n)**2)  # Penaliza alejarse de distribucion uniforme
        return (1.0 - f1) + reg_l2

    np.random.seed(semilla)
    mejor = None
    # 10 inicializaciones: 2 aleatorias + 3 uniformes dispersas + 5 sesgadas
    # Las inicializaciones aleatorias puras tienden a colapsar al mismo minimo local.
    # La mayoria son inicializaciones sesgadas que exploran distintos nodos dominantes.
    # Sin early stop: el optimizador recorre el espacio completo para garantizar convergencia.
    inits_base = [np.random.randn(n) * 0.15 for _ in range(2)]
    inits_unif = [np.ones(n)/n + np.random.randn(n) * dispersa for dispersa in [0.02, 0.06, 0.10]]
    inits_sesg = [np.eye(n)[i % n] * 0.65 + np.ones(n) * 0.12 for i in range(5)]
    inits = inits_base + inits_unif + inits_sesg

    for idx, w0 in enumerate(inits):
        res = minimize(objetivo, w0, method='Nelder-Mead',
                       options={'maxiter': 1500, 'xatol': 1e-5, 'fatol': 1e-5})
        if mejor is None or res.fun < mejor.fun:
            mejor = res
        f1_actual = round(1.0 - mejor.fun, 4)
        num = str(idx + 1).zfill(2)
        print('      init ' + num + '/10 | mejor F1 val = ' + str(f1_actual))

    w_opt = np.exp(mejor.x)
    w_opt = w_opt / w_opt.sum()
    # Garantizar que ningun peso sea menor a 0.05 (piso minimo por nodo)
    w_opt = np.maximum(w_opt, 0.05)
    w_opt = w_opt / w_opt.sum()
    return w_opt


def test_mcnemar(y_true, y_pred_a, y_pred_b):
    """Test de McNemar para comparar dos clasificadores.
    Retorna el estadistico chi2 y el p-valor.
    H0: ambos clasificadores cometen el mismo numero de errores.
    p < 0.05 indica diferencia estadisticamente significativa.
    """
    correct_a = (y_true == y_pred_a)
    correct_b = (y_true == y_pred_b)
    b = np.sum(correct_a & ~correct_b)   # A acierta, B falla
    c = np.sum(~correct_a & correct_b)   # A falla, B acierta
    if b + c == 0:
        return 0.0, 1.0
    chi2  = (abs(b - c) - 1.0)**2 / (b + c)
    from scipy.stats import chi2 as chi2_dist
    p_val = 1 - chi2_dist.cdf(chi2, df=1)
    return chi2, p_val


print('[ OK ] Seccion 3 · Funciones definidas (particion, agregacion, aprendizaje de pesos, McNemar)')
```
